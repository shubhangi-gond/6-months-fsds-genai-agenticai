#### Object-Oriented API

```python
# first create a grid
fig, ax = plt.subplots(2)

#call plot()
ax[0].plot(x1, np.sin(x1), 'b-')
ax[1].plot(x1, np.cos(x1), 'b-')
plt.show()
```

```python
fig = plt.figure()

x2 = np.linspace(0, 5, 10)
y2 = x2 ** 2

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x2, y2, 'r')

axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title')
plt.show();
```

### Error Bar Chart

```python
x9 = np.arange(0, 4, 0.2)
y9 = np.exp(-x9)

e1 = 0.1* np.abs(np.random.randn(len(y9)))
plt.errorbar(x9, y9, yerr = e1, fmt = '.-')

plt.show()
```

```python
A = [15, 30, 45, 22]
B = [15, 25, 50, 20]
z2 = range(4)

plt.bar(z2, A, color = 'b')
plt.bar(z2, B, color = 'r', bottom = A)

plt.show()
```

```python
x15 = np.arange(1,5)
fig, ax = plt.subplots()

ax.plot(x15, x15*1.5)
ax.plot(x15, x15*3.0)
ax.plot(x15, x15/3.0)

ax.legend(['normal', 'fast', 'slow']);
plt.show()
```

```python
x15 = np.arange(1, 5)
fig, ax = plt.subplots()

ax.plot(x15, x15*1.5, label='Normal')
ax.plot(x15, x15*3.0, label='Fast')
ax.plot(x15, x15/3.0, label='Slow')
ax.legend();
plt.show()
```
